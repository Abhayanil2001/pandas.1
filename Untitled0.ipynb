Form a subset of data using Group

```python
df1 = df.groupby(['DATE','location',])[['DATE','location','new_cases','total_deaths','total_vaccinations']].sum().reset_index()
df1
```

```python
df1['new_cases'].replace({'missing_value':0},inplace=True)
df1
```

Will plot subset of specific data

```python
df2=df1[df1['total_deaths']>1000000]
location=df2['location'].unique()
len(location) #finding the total_deaths>1000000 using len
```

```python
location_deaths_greaterthan10000000 = list(df2.location.unique())
location_deaths_greaterthan10000000 #list of people who died
```

```python
import matplotlib.pyplot as plt
for idx in range (0,len(location)):
  C= df2[df2["location"]==location[idx]].reset_index()
  plt.scatter(np.arange(0,len(C)),C['new_cases'],color='blue',label='total_cases')
  plt.scatter(np.arange(0,len(C)),C['total_deaths'],color='red',label='total_deaths')
  plt.scatter(np.arange(0,len(C)),C["total_vaccinations"],color='green',label='total_vaccinations')
  plt.xlabel('number of days since first suspect')
  plt.ylabel("number of cases")
  plt.legend()
plt.show()
```

```python
# make sure df2 is defined properly before running this line
date = df2['DATE'].unique()
df3 = df2.groupby('DATE')[['new_cases','total_deaths']].sum().reset_index()

plt.scatter(df3['DATE'], df3['new_cases'], color='blue', label='new_cases')
plt.scatter(df3['DATE'], df3['total_deaths'], color='red', label='total_deaths')
plt.title('world')
plt.xlabel('date')
plt.ylabel('number of cases')
plt.legend()
plt.show()
```

```python
df4 = df2.groupby(['location'])[['location', 'new_cases',  'total_deaths']].sum().reset_index()
c = df4
plt.scatter(c['location'], c['new_cases'], color='blue', label='new_cases')
plt.scatter(c['location'], c['total_deaths'], color='red', label='total_deaths')
plt.title('World')
plt.xlabel('Location')
plt.ylabel('Number of cases')
plt.show()
```
